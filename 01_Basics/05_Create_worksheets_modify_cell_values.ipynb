````python
from openpyxl.workbook import Workbook
from openpyxl import load_workbook
````

## How to read and change a cell value from an active worksheet in a workbook
1. Read the workbook

````python
wb2 = load_workbook('files/prueba.xlsx')
````

2. Create a new sheet

````python
wb2.create_sheet('Newsheet')
````

3. Select the cell as an index and print its value

````python
active_sheet = wb2.active
cell = active_sheet['A1']
print(cell.value)
````

````python
# another way 
print(wb2.active['B1'].value)
````

3. Change the cell value

````python
active_sheet['A1'] = 'value changed'
````

````python
cell = active_sheet['A1']
print(cell.value)
````

4. Save the changes as a new file

````python
wb2.save('files/modified.xlsx')
````
